```python
import math
from CoolProp.CoolProp import PropsSI
```

```python
# Data
H = 1.5          # m
W = 3.0          # m
L = 0.03         # m  
theta_deg = 20.0
eps1 = 0.8
eps2 = 0.9

T_hot_C = 80.0
T_cold_C = 32.0

sigma = 5.67e-8  # W/m^2 K^4

# Geometría y temperaturas
A = H * W
theta = math.radians(theta_deg)

T_hot = T_hot_C + 273.15
T_cold = T_cold_C + 273.15
T_avg = 0.5 * (T_hot + T_cold)
dT = T_hot - T_cold

# Propiedades del aire con CoolProp
P = 101325.0  # Pa - 1atm

rho = PropsSI("D", "T", T_avg, "P", P, "Air")   # densidad
mu = PropsSI("V", "T", T_avg, "P", P, "Air")    # viscosidad dinámica
k = PropsSI("L", "T", T_avg, "P", P, "Air")     # conductividad térmica
cp = PropsSI("C", "T", T_avg, "P", P, "Air")    # calor específico

beta = 1.0 / T_avg
nu = mu / rho
alpha = k / (rho * cp)
Pr = nu / alpha

# Rayleigh
Ra = 9.81 * beta * dT * L**3 / (nu * alpha)

# Nu
Ra_cos = Ra * math.cos(theta)

term1 = max(0.0, 1.0 - 1708.0 / Ra_cos)
term2 = 1.0 - 1708.0 * (math.sin(1.8 * theta) ** 1.6) / Ra_cos
term3 = max(0.0, (Ra_cos / 5830.0) ** (1.0 / 3.0) - 1.0)

Nu = 1.0 + 1.44 * term1 * term2 + term3

# Convección
h = k * Nu / L
Q_conv = h * A * dT

# Radiación
den_rad = 1.0 / eps1 + 1.0 / eps2 - 1.0
Q_rad = sigma * A * (T_hot**4 - T_cold**4) / den_rad

Q_total = Q_conv + Q_rad
```

```python
# Resultados
# -----------------------------
print(f"T_avg = {T_avg - 273.15:.2f} °C")
print(f"rho = {rho:.5f} kg/m^3")
print(f"mu = {mu:.6e} Pa·s")
print(f"k = {k:.6f} W/m·K")
print(f"cp = {cp:.2f} J/kg·K")
print(f"nu = {nu:.6e} m^2/s")
print(f"alpha = {alpha:.6e} m^2/s")
print(f"Pr = {Pr:.4f}")
print(f"Ra = {Ra:.4e}")
print(f"Nu = {Nu:.4f}")
print(f"h = {h:.4f} W/m^2·K")
print(f"Q_conv = {Q_conv:.2f} W")
print(f"Q_rad = {Q_rad:.2f} W")
print(f"Q_total = {Q_total:.2f} W")
```

```python
import math
from CoolProp.CoolProp import PropsSI

# Datos del problema

L = 5.0           # longitud del tubo (m)
D = 0.01          # diametro (m)
P = 750e3         # presión (Pa)
Tw = 28 + 273.15  # temperatura pared (K)

g = 9.81


# Propiedades de saturación

Tsat = PropsSI("T","P",P,"Q",0,"R134a")

rho_l = PropsSI("D","P",P,"Q",0,"R134a")
rho_v = PropsSI("D","P",P,"Q",1,"R134a")

h_l = PropsSI("H","P",P,"Q",0,"R134a")
h_v = PropsSI("H","P",P,"Q",1,"R134a")

h_fg = h_v - h_l

mu_l = PropsSI("V","P",P,"Q",0,"R134a")
mu_v = PropsSI("V","P",P,"Q",1,"R134a")

k_l = PropsSI("L","P",P,"Q",0,"R134a")

#Nu
dT = Tsat - Tw

h = 0.555*((g*rho_l*(rho_l-rho_v)*h_fg*k_l**3)/(mu_l*D*dT))**0.25
A = math.pi*D*L
Q = h*A*dT
m_dot = Q/h_fg


A_flow = math.pi*D**2/4

V_in = m_dot/(rho_v*A_flow)



Re = rho_v*V_in*D/mu_v

```

```python

print("Temperatura de saturación:", Tsat-273.15, "C")
print("Coeficiente convectivo h:", h, "W/m^2K")
print("Tasa de calor Q:", Q, "W")
print("Tasa de condensación m_dot:", m_dot, "kg/s")
print("Velocidad de entrada:", V_in, "m/s")
print("Reynolds entrada:", Re)
```
