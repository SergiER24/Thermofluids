```python
import math
from CoolProp.CoolProp import PropsSI

P = 4.25e3          # Pa
Ts_C = 20.0         # °C
Ts = Ts_C + 273.15  # K

D = 0.03            # m
L = 8.0             # m
n_total = 100       # tubos totales
N_rows = 10         # filas verticales
g = 9.81            # m/s^2

fluid = 'IF97::Water'   # 'Water'

# Saturación a la presión dada
T_sat = PropsSI('T', 'P', P, 'Q', 1, fluid)   # K, vapor saturado
T_sat_C = T_sat - 273.15

# Temperatura de película
T_f = 0.5 * (T_sat + Ts)
T_f_C = T_f - 273.15

# Propiedades
# Vapor saturado a P
rho_v = PropsSI('D', 'P', P, 'Q', 1, fluid)   # kg/m^3
h_v   = PropsSI('H', 'P', P, 'Q', 1, fluid)   # J/kg

# Líquido saturado a P (para h_fg)
h_l_sat = PropsSI('H', 'P', P, 'Q', 0, fluid) # J/kg
h_fg = h_v - h_l_sat                          # J/kg

# Líquido condensado a T_f
rho_l = PropsSI('D', 'P', P, 'T', T_f, fluid) # kg/m^3
mu_l  = PropsSI('V', 'P', P, 'T', T_f, fluid) # Pa·s
k_l   = PropsSI('L', 'P', P, 'T', T_f, fluid) # W/m·K
cp_l  = PropsSI('C', 'P', P, 'T', T_f, fluid) # J/kg·K


dT = T_sat - Ts

h_fg_star = h_fg + 0.68 * cp_l * dT #calor latente

# h para el primer tubo
h1 = 0.729 * (
    g * rho_l * (rho_l - rho_v) * h_fg_star * (k_l ** 3) 
    / (mu_l * dT * D)
) ** 0.25

# h del último tubo/fila N y lista por fila
h_rows = [h1 / (j ** 0.25) for j in range(1, N_rows + 1)]
h_bank = h1 / (N_rows ** 0.25)

# Transferencia de calor total
A_total = n_total * math.pi * D * L
Q_dot = h_bank * A_total * dT

# Tasa de condensación
m_dot = Q_dot / h_fg_star

# Resultados

print(f"T_sat = {T_sat_C:.3f} °C")
print(f"T_f   = {T_f_C:.3f} °C")

print("\nPropiedades:")
print(f"rho_v   = {rho_v:.6f} kg/m^3")
print(f"rho_l   = {rho_l:.3f} kg/m^3")
print(f"mu_l    = {mu_l:.6e} Pa·s")
print(f"k_l     = {k_l:.6f} W/m·K")
print(f"cp_l    = {cp_l:.3f} J/kg·K")
print(f"h_fg    = {h_fg:.3f} J/kg")
print(f"h_fg*   = {h_fg_star:.3f} J/kg")
```

```python
print("\nCoeficientes:")
print(f"h_1     = {h1:.3f} W/m^2·K")
for j, h_j in enumerate(h_rows, start=1):
    print(f"h_fila_{j} = {h_j:.3f} W/m^2·K")
print(f"h_banco = {h_bank:.3f} W/m^2·K")

print("\nTransferencia:")
print(f"A_total = {A_total:.6f} m^2")
print(f"Q_dot   = {Q_dot:.3f} W")
print(f"m_dot   = {m_dot:.6f} kg/s")
```
