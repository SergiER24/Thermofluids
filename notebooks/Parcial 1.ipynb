```python
import math
from CoolProp.CoolProp import PropsSI
import numpy as np
H = 1.5
W = 3.0
L = 0.03
theta = np.deg2rad(20)
T_hot = 80 + 273.15
T_cold = 32 + 273.15

A = H * W
dT = 80 - 32
Tf = (80 + 32) / 2 + 273.15

k = 0.027786
nu = 1.856e-5 #mu
Pr = 0.72124
beta = 1 / Tf
g = 9.81

Gr = g * beta * dT * L**3 / nu**2
Ra = Gr * Pr

Ra_cos = Ra * math.cos(theta)

term1 = 1 - 1708 / Ra_cos
term2 = 1 - 1708 * (math.sin(1.8 * theta) ** 1.6) / Ra_cos
term3 = (Ra_cos / 5830) ** (1/3) - 1

Nu = 1 + 1.44 * term1 * term2 + term3

h = k * Nu / L
Qconv = h * A * dT

sigma = 5.67e-8
eps1 = 0.8
eps2 = 0.9

den = 1/eps1 + 1/eps2 - 1
Qrad = sigma * A * (T_hot**4 - T_cold**4) / den

Qtotal = Qconv + Qrad

print(Ra)
print(Nu)
print(Qconv)
print(Qrad)
print(Qtotal)
```

```python

L = 5.0
D = 0.01
P = 750e3
Tw = 28.0 + 273.15
g = 9.81

Tsat = PropsSI("T", "P", P, "Q", 0, "R134a")
rho_l = PropsSI("D", "P", P, "Q", 0, "R134a")
rho_v = PropsSI("D", "P", P, "Q", 1, "R134a")

h_g = PropsSI("H", "P", P, "Q", 1, "R134a")
h_out = PropsSI("H", "P", P, "T", Tw, "R134a")

mu_l = PropsSI("V", "P", P, "Q", 0, "R134a")
mu_v = PropsSI("V", "P", P, "Q", 1, "R134a")
k_l = PropsSI("L", "P", P, "Q", 0, "R134a")
cp_l = PropsSI("C", "P", P, "Q", 0, "R134a")

dT = Tsat - Tw
hfg = h_g - PropsSI("H", "P", P, "Q", 0, "R134a")
hfg_star = hfg * (1.0 + 0.68 * cp_l * dT / hfg)

h = 0.555 * ((g * rho_l * (rho_l - rho_v) * hfg_star * k_l**3) / (mu_l * D * dT))**0.25

A = math.pi * D * L
Qdot = h * A * dT

mdot = Qdot / (h_g - h_out)

A_flow = math.pi * D**2 / 4.0
V_in = mdot / (rho_v * A_flow)

Re_in = rho_v * V_in * D / mu_v

print(Tsat - 273.15)
```
